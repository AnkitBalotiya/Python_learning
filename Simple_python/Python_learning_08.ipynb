```python
# Small excerise - analysing differnt carrer field in data and there salary (median)

import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import numpy as np

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df.info()
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False)
```

```python
df.groupby('job_title_short')['salary_year_avg'].median().sort_values().plot(kind='barh')
```
