```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df =  dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
x = np.arange(0,5,0.1)
y = np.sin(x)
plt.plot(x,y)
```

```python
x = [1,2,3,4,5,6]
y = [1,2,1,4,12,2]
plt.plot(x,y)
```

```python
date_counts = df['job_posted_date'].value_counts().sort_index()
date_counts
```

```python
plt.plot(date_counts.index ,date_counts)

df['job_posted_month'] = df['job_posted_date'].dt.month
df['job_posted_month']
```

```python

monthly_counts = df[df['job_location']=='India']['job_posted_month'].value_counts()
monthly_counts.sort_index(inplace=True)
print(monthly_counts)
plt.plot(monthly_counts.index,monthly_counts)
```

```python
job_title = df[df['job_location']=='India'].job_title_short.value_counts().sort_values(ascending=True).head(5)
plt.barh(job_title.index , job_title)
job_title
plt.title('Job title posting')
plt.xlabel('Counts of Jobs Posting')
```

```python
job_title = df[df['job_location']=='India'].job_title_short.value_counts().sort_values(ascending=False)
plt.bar(job_title.index , job_title)
# job_title
plt.title('Job title posting')
plt.ylabel('Counts of Jobs Posting')
plt.xticks(rotation=45,ha='right')
plt.show()
```
